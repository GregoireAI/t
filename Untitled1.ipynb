```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, lambdify, sympify

def plot_function(function_str, x_range):
    x_symbol = symbols('x')
    try:
        # Parse the function string into a symbolic expression
        function = sympify(function_str)
        # Turn the symbolic expression into a callable function
        func = lambdify(x_symbol, function, 'numpy')

        # Generate x values
        x_values = np.linspace(x_range[0], x_range[1], 400)
        # Generate y values
        y_values = func(x_values)

        # Plotting the function
        plt.figure(figsize=(8, 6))
        plt.plot(x_values, y_values, label=function_str)
        plt.xlabel('x')
        plt.ylabel('f(x)')
        plt.title('Graph of the function')
        plt.legend()
        plt.grid(True)
        plt.show()
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
function_input = input("Enter a function (e.g., 'x**2 + 3*x - 5'): ")
plot_function(function_input, [-10, 10])
```
